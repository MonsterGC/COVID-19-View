```python
# encoding=utf-8
from pandas import DataFrame,Series
import pandas as pd
from matplotlib import pyplot
import matplotlib.pyplot as plt

# 解决乱码问题
plt.rcParams["font.sans-serif"]=["SimHei"]
plt.rcParams["axes.unicode_minus"]=False
```

```python
data = pd.read_excel("全国新冠状肺炎数据.xls",encoding='gbk')
data = data.fillna(0)
# 前几条数据
data.head()
```

```python
squares=data['确诊(新增）']
input_value=range(len(squares))
#plt.plot(squares)
plt.plot(input_value,squares,linewidth=1)
 
#设置图表标题，并给坐标轴加标签
plt.title("确诊(新增）-全国",fontsize=12)
plt.xlabel("时间(/天)",fontsize=14)
plt.ylabel("确诊(新增）",fontsize=14)
 
#设置刻度标记的大小
plt.tick_params(axis='both',labelsize=14)
plt.show()
```
