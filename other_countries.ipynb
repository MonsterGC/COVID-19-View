```python
# encoding=utf-8
from pandas import DataFrame,Series
import pandas as pd
from matplotlib import pyplot
import matplotlib.pyplot as plt

# 解决乱码问题
plt.rcParams["font.sans-serif"]=["SimHei"]
plt.rcParams["axes.unicode_minus"]=False
```

```python
data = pd.read_excel("全球主要国家统计数据.xlsx",encoding='gbk')
data = data.fillna(0)
# 前几条数据
data.head()
```

```python
# DataFrame
data = pd.DataFrame(data)
print(data)
```

```python
# 更具省份名称求均值
data.groupby('国家/地区').mean()
```

```python
# 分组
provinces_data = dict()
for key,name in data.groupby('国家/地区'):
    provinces_data[key] = name
```

```python
# 分组出来的数据
provinces_data['美国']
```

```python
# 获取前8条
filter_key = data.groupby('国家/地区').mean()["累计死亡人数"].sort_values(ascending=False)[0:8]

for tmp in provinces_data:
    if tmp in filter_key:
        x = range(0, len(provinces_data[tmp]['累计死亡人数']))
#     label = lambda label:'' if tmp not in filter_key else tmp
        plt.plot(x, provinces_data[tmp]['累计死亡人数'] ,  mec='r', mfc='w', label= tmp)
plt.legend()  # 让图例生效
plt.xlabel('时间(/天)') #X轴标签
plt.ylabel("累计死亡人数-均值最大前8") #Y轴标签
plt.show()
```
