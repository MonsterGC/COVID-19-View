```python
# encoding=utf-8
from pandas import DataFrame,Series
import pandas as pd
from matplotlib import pyplot
import matplotlib.pyplot as plt

# 解决乱码问题
plt.rcParams["font.sans-serif"]=["SimHei"]
plt.rcParams["axes.unicode_minus"]=False
```

```python
data = pd.read_excel("全国重点省市统计数据.xlsx",encoding='gbk')
data = data.fillna(0)
# 前几条数据
data.head()
```

```python
# DataFrame
data = pd.DataFrame(data)
print(data)
```

```python
# 更具省份名称求均值
data.groupby('省市名称').mean()
```

```python
# 分组
provinces_data = dict()
for key,name in data.groupby('省市名称'):
    provinces_data[key] = name
```

```python
# 分组出来的数据
provinces_data['上海']
```

```python
for tmp in provinces_data:
    x = range(0, len(provinces_data[tmp]['确诊人数']))
    label = lambda label:'' if tmp not in ['湖北省' , '武汉'] else tmp
    plt.plot(x, provinces_data[tmp]['确诊人数'] ,  mec='r', mfc='w', label=label(2))
plt.legend()  # 让图例生效
plt.xlabel('时间(/天)') #X轴标签
plt.ylabel("确诊人数") #Y轴标签
plt.show()
```

```python
# 含武汉、湖北
for tmp in provinces_data:
    x = range(0, len(provinces_data[tmp]['治愈人数']))
    label = lambda label:'' if tmp not in ['湖北省' , '武汉'] else tmp
    plt.plot(x, provinces_data[tmp]['治愈人数'] ,  mec='r', mfc='w', label=label(2))
plt.legend()  # 让图例生效
plt.xlabel('时间(/天)') #X轴标签
plt.ylabel("治愈人数") #Y轴标签
plt.show()
```
